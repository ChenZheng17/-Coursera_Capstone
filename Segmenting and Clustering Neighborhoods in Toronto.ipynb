```python
import sys
if sys.version_info[0] == 2:
    from urllib2 import urlopen  # Python 2
else:
    from urllib.request import urlopen  # Python3

html = urlopen("https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M")
```

```python
import requests
```

```python
soup = BeautifulSoup(html, 'lxml')
```

```python
match=soup.tbody.text
print(match)
```

```python
match1=match.split('\n')
match1
```

# # Remove the ‘/n’ 

```python
   while '' in match1:
        match1.remove('')
print (match1)
```

```python
step = 3
match2 = [match1[i:i+step] for i in range(0,len(match1),step)]
print(match2)
```

```python
import pandas as pd
df = pd.DataFrame(match2, columns=['Postcode', 'Borough', 'Neighbourhood'])
df.head()
```

```python
df=df.drop(labels=None,axis=0, index=0, columns=None, inplace=False)
df.head()
```

```python
df=df[ ~ df['Borough'].str.contains('Not assigned')]
df.head()
```

```python
df1=df.drop_duplicates('Postcode')
df1.head()
```

```python
df2=df1.drop(['Neighbourhood'],axis=1).set_index('Postcode')
df2.head()
```

```python
df3 = pd.DataFrame(df.groupby('Postcode')['Neighbourhood'].apply(lambda x: "%s" % ', '.join(x)))
df3.head()
```

```python
df4=df2.join([df3],how='outer')
df4.head()
```

```python
df4.index.name = 'Postcode'
df4.head()
```

# # The final dataframe

```python
df5=df4.reset_index('Postcode')
df5.head()
```
